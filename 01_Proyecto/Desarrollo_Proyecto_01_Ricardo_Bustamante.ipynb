#### Ejercicio N°1:


Escribir un programa que pregunte al usuario o genere las ventas de los últimos 5 días, y muestre por pantalla una serie (Dataframe) con los datos de las ventas por producto (opcional): Nro de unidades y precio unitario, indexada por 
los días, antes y después de aplicarles un descuento del 5%.

```python
np.random.seed(147258369)

filas = 5

data = {
    'Unidades': np.random.randint(100, size=(filas)),
    'Precio_unitario': [0.75] * filas
    }
```

```python
df_ventas = pd.DataFrame(
    data = data,
    index = pd.date_range(end = "20221115", periods = filas),
    )
```

```python
df_ventas 
```

```python
df_ventas['Total_venta_bruta'] = df_ventas['Unidades'] * df_ventas['Precio_unitario']
```

```python
df_ventas['Descuento'] = (df_ventas['Total_venta_bruta'] * 0.05).round(2)
```

```python
df_ventas['Total_venta_neta'] = df_ventas['Total_venta_bruta'] - df_ventas['Descuento']
```

```python
df_ventas
```

#### Ejercicio Nº 3:

Escribir programa que genere y muestre por pantalla un DataFrame con los datos de la siguiente tabla:

|Mes|Ventas|Gastos|
|---|------|------|
|Enero|30500|22000|
|Febrero|35600|23450|
|Marzo|28300|18100|
|Abril|33900|35700|
|Mayo|42500|32450|

A continuación, generar nuevas columna:
1. Columna 1 = Gasto/Ventas*100.
2. Columna 2 = Si Ventas>= 30000, imprimir “Meta Superada”, caso contrario imprimir “Meta no Superada”.
3. Columna 3 = Si Gasto>Ventas, calcular el sobre gasto.
4. Obtener el total de las ventas de los últimos 3 meses
5. Obtener el total de gastos de febrero a abril

```python
df_ejer3 = pd.DataFrame(
    data = {'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo'],
            'Ventas': [30500, 35600, 28300, 33900, 42500],
            'Gastos': [22000, 23450, 18100, 35700, 32450]}
)
```

```python
df_ejer3
```

*> Desarrollo punto 1). Agregar la columna 1*

```python
df_ejer3['% Gastos/Ventas'] = ((df_ejer3['Gastos'] / df_ejer3['Ventas']) * 100).round(2)
```

*> Desarrollo punto 2). Agregar la columna 2*

```python
df_ejer3['Meta'] = np.where(df_ejer3.Ventas >= 30000, 'Meta Superada', 'Meta no Superada')
```

*> Desarrollo punto 3). Agregar la columna 3*

```python
df_ejer3['Sobre gastos'] = list(map(lambda x: x if x > 0 else 0, list(df_ejer3['Gastos'] - df_ejer3['Ventas'])))
```

```python
df_ejer3
```

*> Desarrollo punto 4).*

```python
df_ejer3.iloc[2:, :2].sum()
```

*> Desarrollo punto 5).*

```python
df_ejer3.iloc[1:4, [0,2]].sum()
```
