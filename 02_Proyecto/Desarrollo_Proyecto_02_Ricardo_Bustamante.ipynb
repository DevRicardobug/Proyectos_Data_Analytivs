#### Ejercicio Nº 3

Crear una función que permita generar 3 variables cualitativas ordinales a partir de las variables numéricas con los siguientes criterios:

* Si es mayor a la “media + 1*desviación estándar” → Alto
* Si está entre “media - 1*desviación estándar” y “media + 1*desviación estándar” → Medio
* Si es menor a la “media - 1*desviación estándar” → Bajo

Nota: Mostrar con las nuevas variables generadas.

Elegir con libertad las variables a trabajar.

*La función debe contar con su respectiva documentación.*

```python
col_evaluar = df_autos.criterio4
```

```python
def categoria (x:float):
    '''
    Devuelve la clasificacion cualitativa de una variable definida en un DataFrame.

    * Si es mayor a la “media + 1*desviación estándar” → Alto
    * Si está entre “media - 1*desviación estándar” y “media + 1*desviación estándar” → Medio
    * Si es menor a la “media - 1*desviación estándar” → Bajo
    
    Parámetros
    ----------
    x : float
    '''

    media = col_evaluar.mean()
    desv = col_evaluar.std()

    if x > (media + 1*desv):
        return 'Alto'
    elif (media + 1*desv) > x and x > (media - 1*desv):
        return 'Medio'
    elif x < (media - 1*desv):
        return 'Bajo'
```

```python
df_autos['Media'] = col_evaluar.apply(categoria)
```

#### Ejercicio Nº 4

Crear una función que muestre los siguientes gráficos (simultáneamente) para las variables numéricas, según una variable cualitativa:

* Cajas y bigotes
* Barras de la media
* Histograma con densidad (Curva)

*La función debe contar con su respectiva documentación.*

```python
def graficos (dataframe, v_cualitativa:str, v_cuantitativa:str):
    '''
    Devuele 3 graficos de un DataFrame.

    * Boxplot
    * Barplot
    * Histplot

    Parámetros
    ----------
    dataframe : DataFrame
    v_cualitativa : Nombre de la variable cualitativa del Dataframe
    v_cuantitativa : Nombre de la variable cuantitativa del Dataframe
    '''
    l_v_cualitativa = np.array(dataframe[v_cualitativa])
    l_v_cuantitativa = np.array(dataframe[v_cuantitativa])
    u_listas = list(zip(l_v_cualitativa, l_v_cuantitativa))

    usa = []
    europe = []
    japan = []

    for i in u_listas:
        if i[0] == 'USA':
            usa.append(i[1])
        elif i[0] == 'Europe':
            europe.append(i[1])
        elif i[0] == 'Japan':
            japan.append(i[1])
    
    consolidacion = [usa, europe, japan]

    fig, (axes1,axes2,axes3) = plt.subplots(nrows=1, ncols=3, figsize=(22,8))

    axes1.boxplot(x=consolidacion, patch_artist=True, medianprops={"color": "white", "linewidth": 2}, boxprops={"facecolor": "#E02B4C", "edgecolor": "#E02B4C","linewidth": 2}, whiskerprops={"color": "#E02B4C", "linewidth": 2}, capprops={"color": "#E02B4C", "linewidth": 2})
    axes1.set_title(f'Grafico_01 - Cajas y Bigotes\n {v_cuantitativa} - {v_cualitativa}', fontsize="x-large")
    axes1.set_xlabel('1:USA    2:Europe    3:Japan', backgroundcolor="#E02B4C", color="white", fontsize="large")
    axes1.set_ylabel(v_cuantitativa, fontsize="large")

    axes2.bar(l_v_cualitativa, l_v_cuantitativa, color="#8267E0")
    axes2.set_title(f'Grafico_02 - Barras\n {v_cuantitativa} - {v_cualitativa}', fontsize="x-large")
    axes2.set_xlabel(v_cualitativa, fontsize="large")
    axes2.set_ylabel(v_cuantitativa, fontsize="large")

    axes3.hist(l_v_cuantitativa, color="#58B9E0", linewidth=0.5, edgecolor="white")
    axes3.set_title(f'Grafico_03 - Histograma\n {v_cuantitativa}', fontsize="x-large")
    axes3.set_xlabel(v_cuantitativa, fontsize="large")


    fig.tight_layout() #mejora la visualización de los ejes de cada gráfico

    plt.show()
```

```python
graficos(dataframe=df_autos, v_cualitativa='Pais_Origen', v_cuantitativa='Desplazamiento')
```

#### Ejercicio Nº 5

Crear una función que devuelva un mapa de calor con la correlación de todas las variables numéricas (incluye las generadas), para cada país.

*La función debe contar con su respectiva documentación.*

```python
def headmap (dataframe):
    '''
    Devuelve el heatmap de un DataFrame

    Parámetros
    ----------
    dataframe : DataFrame
    '''
    plt.figure(figsize=(25,10))
    sns.heatmap(dataframe.corr(),annot=True);
```

```python
headmap(df_autos)
```
